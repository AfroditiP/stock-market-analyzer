```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
# Create a database connection
conn = sqlite3.connect("../data/stocks.db")

# Load all stocks into the database
tickers = ["AAPL", "TSLA", "MSFT", "GOOGL", "AMZN"]

for ticker in tickers:
    df = pd.read_csv(f"../data/{ticker}.csv", index_col=0, parse_dates=True, skiprows=[1,2])
    df.to_sql(ticker, conn, if_exists="replace", index=True)
    print(f"{ticker} loaded into database!")

print("\nDatabase ready!")
```

```python
query = """
SELECT *
FROM AAPL
LIMIT 5
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT Price, Close, Volume
FROM AAPL
WHERE Close > 200
ORDER BY Close DESC
LIMIT 10
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT Price, Close, Volume
FROM AAPL
WHERE Close < 70
ORDER BY Close ASC
LIMIT 10
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT
    strftime('%Y', Price) as Year,
    ROUND(AVG(Close), 2) as Avg_Price,
    ROUND(MAX(Close), 2) as Max_Price,
    ROUND(MIN(Close), 2) as Min_Price
FROM AAPL
GROUP BY Year
ORDER BY Year
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT
    a.Price as Date,
    ROUND(a.Close, 2) as AAPL_Price,
    ROUND(t.Close, 2) as TSLA_Price
FROM AAPL a
JOIN TSLA t ON a.Price = t.Price
ORDER BY a.Price
LIMIT 10
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT
    ROUND(MAX(Close), 2) as Max_Price,
    ROUND(MIN(Close), 2) as Min_Price,
    ROUND(AVG(Close), 2) as Avg_Price
FROM TSLA
"""

result = pd.read_sql(query, conn)
print(result)
```
