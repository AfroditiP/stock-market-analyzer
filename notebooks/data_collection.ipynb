```python
tickers = ["AAPL", "TSLA", "MSFT", "GOOGL", "AMZN"]
for ticker in tickers:
    print(f"Downloading {ticker}...")
    df = yf.download(ticker, start="2020-01-01", end="2024-12-31")
    df.to_csv(f"../data/{ticker}.csv")
    print(f"{ticker} saved!")
print("\n All done!")
```
